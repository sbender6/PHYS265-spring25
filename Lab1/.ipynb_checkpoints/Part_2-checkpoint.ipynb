# Part 2: The Gravitational Potential of the Earth-Moon System

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.cm as cm
from matplotlib.colors import LogNorm

G = 6.67*10**-11 
Me = 5.9*10**24
Mm = 7.3*10**22
dem = 3.8*10**8
```

### 1. The 2D color-mesh plot of the gravitational potential of the Earth-Moon system

```python
def egp(M, xM, yM, x, y):
    r = np.sqrt((x-xM)**2+(y-yM)**2) 
    
    return (-G*M)/r

def mgp(M, xM, yM, x, y): 
    r = np.sqrt((x-xM)**2+(y-yM)**2) 

    return -(G*M)/r
mx = dem/np.sqrt(2) 
my = dem/np.sqrt(2) 

fig, ax = plt.subplots() 

X = np.linspace(-1.5*dem, 1.5*dem, 100) 
Y = np.linspace(-1.5*dem, 1.5*dem, 100) 

x_grid, y_grid = np.meshgrid(X, Y) 

earth = np.abs(egp(Me, 0, 0, x_grid, y_grid))
moon = np.abs(mgp(Mm, mx, my, x_grid, y_grid)) 
C = earth + moon

colormesh = ax.pcolormesh(x_grid, y_grid, C, cmap=cm.jet, norm=LogNorm())
ax.set_aspect('equal') 

cbar = fig.colorbar(colormesh)

ax.set_title("The Gravitational Potential of the Earth-Moon System (As Singularities)") 
ax.set_xlabel("x distance (m)") 
ax.set_ylabel("y distance (m)") 
cbar.set_label("Gravitational potential (J/kg)")
```

### 2. The 2D contour plot of the gravitational potential of the Earth-Moon system

```python
fig, ax = plt.subplots() 

levels = np.logspace(np.log10(np.min(C[C > 0])), np.log10(np.max(C)), num = 10) 
contour = ax.contour(x_grid, y_grid, C, levels = levels, cmap=cm.jet, norm=LogNorm()) 
ax.set_aspect('equal') 

cbar = fig.colorbar(contour)

ax.set_title("The Gravitational Potential of the Earth-Moon System (As Singularities)") 
ax.set_xlabel("x distance (m)") 
ax.set_ylabel("y distance (m)") 
cbar.set_label("Gravitational potential (J/kg)")
```
