# Part 1: The Gravitational Potential of the Earth

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.cm as cm

G = 6.67*10**-11 
Me = 5.9*10**24
dem = 3.8*10**8
```

### 1. The function of gravitational potential at some evaluation point

```python
def grav_potential(M, xM, yM, x, y):
    r = np.sqrt((x-xM)**2+(y-yM)**2) 
    
    return (-G*M)/r
```

### 2. The 1D plot of gravitational potential v. distance from the Earth

```python
fig, ax = plt.subplots() 

x1 = np.linspace(0, 1.5*dem, 100) 
ax.set_yscale('log')  
ax.plot(x1, np.abs(grav_potential(Me, 0, 0, x1, 0)))

ax.set_title("Gravitational Potential v. Distance from Earth") 
ax.set_ylabel("Absolute value of gravitational potential (m\u00b2/s\u00b2)") 
ax.set_xlabel("Distance from the Earth (m)") 
```

### 3. The 2D color-mesh plot of the gravitational potential v. distance from Earth

```python
from matplotlib.colors import LogNorm

fig, ax2 = plt.subplots() 

X = np.linspace(-1.5*dem, 1.5*dem, 100) 
Y = np.linspace(-1.5*dem, 1.5*dem, 100) 

x_grid, y_grid = np.meshgrid(X, Y) 

C = np.abs(grav_potential(Me, 0, 0, x_grid, y_grid)) 

colormesh = ax2.pcolormesh(x_grid, y_grid, C, cmap=cm.jet, norm=LogNorm())
ax2.set_aspect('equal') 

cbar = fig.colorbar(colormesh)

ax2.set_title("The Gravitational Potential of the Earth (As a Singularity)") 
ax2.set_xlabel("x distance (m)") 
ax2.set_ylabel("y distance (m)") 
cbar.set_label("Gravitational potential (J/kg)") 
```
