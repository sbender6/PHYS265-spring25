# Part 3: The Gravitational Force Field of the Earth-Moon System

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.cm as cm
from matplotlib.colors import LogNorm

G = 6.67*10**-11 
g = 9.81
Me = 5.9*10**24
Mm = 7.3*10**22
Ma = 5500
Re = 6378*10**3
Rm = 1737*10**3
dem = 3.8*10**8
ve = 2.4*10**3
m = 1.3*10**4  
```

### 1. The function of gravitational force

```python
def F(M1, m2, x1, y1, x2, y2):  
    r = np.sqrt((x2-x1)**2+(y2-y1)**2) 
    F21 = -G*((M1*m2)/r**2)
    F21 = np.where(r<Re, np.nan, F21) 
    F21x = F21*(x2-x1)/r
    F21y = F21*(y2-y1)/r

    return F21x, F21y
```

### 2. 2D streamplot of the gravitational force asserted on the Apollo 11 command module by the Earth-Moon system

```python
ny = 100
nx = 100
y = np.linspace(-1.5*dem, 1.5*dem, ny) 
x = np.linspace(-1.5*dem, 1.5*dem, nx) 

fig, ax = plt.subplots() 

x_grid, y_grid = np.meshgrid(x, y) 

mx = dem/np.sqrt(2) 
my = dem/np.sqrt(2) 

Ef = F(Me, Ma, 0, 0, x_grid, y_grid)
Mf = F(Mm, Ma, mx, my, x_grid, y_grid)
Fx = Ef[0]+Mf[0]
Fy = Ef[1]+Mf[1]

colors = np.sqrt(Fx**2+Fy**2) 
stream = ax.streamplot(x_grid, y_grid, Fx, Fy, color=colors, cmap = cm.jet, norm=LogNorm()) 
cbar = fig.colorbar(stream.lines) 

ax.set_aspect('equal') 
ax.set_title("The Gravitational Force of the Earth-Moon System") 
ax.set_xlabel("x (m)") 
ax.set_ylabel("y (m)") 
cbar.set_label("Gravitational force (N)") 
```
