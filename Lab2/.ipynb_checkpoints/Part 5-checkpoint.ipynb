## Part 5: A non-uniform Earth 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy
from scipy.integrate import solve_ivp 
from scipy.integrate import quad
```

```python
G = 6.6743*10**-11 # gravitational constant [m^3/kg/s^2]
g0 = 9.81 # gravitational acceleration [m/s^2]
Me = 5.972*10**24 # mass of the earth [kg] 
Mm = 7.35*10**22 # mass of the moon [kg] 
Re = 6378.1e3 # radius of the earth [m] 
Rm = 1738.1e3 # radius of the moon [m] 
w = 7.272*10**-5 # earth's rotation rate at the equator [rad/s] 
```

### 1. 

```python
pn = 1
n = [0, 1, 2, 9]

def p0(y): 
    pr = pn*(1-y**2/Re**2)**n[0]
    return pr
def p1(y): 
    pr = pn*(1-y**2/Re**2)**n[1]
    return pr
def p2(y): 
    pr = pn*(1-y**2/Re**2)**n[2]
    return pr
def p9(y): 
    pr = pn*(1-y**2/Re**2)**n[3]
    return pr

fig, ax = plt.subplots() 
d = np.linspace(0, Re, 1000)
ax.plot(d, p0(d), color='indianred', label="n=0") 
ax.plot(d, p1(d), color='darkorange', label="n=1") 
ax.plot(d, p2(d), color='gold', label="n=2") 
ax.plot(d, p9(d), color='yellowgreen', label="n=9") 
ax.set_title("Figure 5: Normalized Density Profile as a Function of Radius for \nDifferent n's") 
ax.set_xlabel("Distance from the center of the Earth (m)") 
ax.set_ylabel("Density profile (kg/m\u00b3)") 
ax.legend(); 
```

Figure 5: This graph models the change in the density as an object gets further from the center of the Earth, at different density constants.

### 2.  

```python
pr0 = lambda r: (1-(r/Re)**2)**n[0]
pr1 = lambda r: (1-(r/Re)**2)**n[1] 
pr2= lambda r: (1-(r/Re)**2)**n[2] 
pr9 = lambda r: (1-(r/Re)**2)**n[3]

pn0 = Me/(4*np.pi*scipy.integrate.quad((lambda r: pr0(r)*r**2), 0, Re)[0])
pn1 = Me/(4*np.pi*scipy.integrate.quad((lambda r: pr1(r)*r**2), 0, Re)[0])
pn2 = Me/(4*np.pi*scipy.integrate.quad((lambda r: pr2(r)*r**2), 0, Re)[0])
pn9 = Me/(4*np.pi*scipy.integrate.quad((lambda r: pr9(r)*r**2), 0, Re)[0])

def F0(y): 
    M = pn0*4*np.pi*scipy.integrate.quad((lambda r: pr0(r)*r**2), 0, y)[0]
    F = G*M/y**2
    return F
def F1(y): 
    M = pn1*4*np.pi*scipy.integrate.quad((lambda r: pr1(r)*r**2), 0, y)[0]
    F = G*M/y**2
    return F
def F2(y): 
    M = pn2*4*np.pi*scipy.integrate.quad((lambda r: pr2(r)*r**2), 0, y)[0]
    F = G*M/y**2
    return F
def F9(y): 
    M = pn9*4*np.pi*scipy.integrate.quad((lambda r: pr9(r)*r**2), 0, y)[0]
    F = G*M/y**2
    return F

d = np.linspace(0.01, Re, 1000) 

F0_plot = [] 
F1_plot = [] 
F2_plot = [] 
F9_plot = [] 

for i in range(len(d)): 
    if d[i] ==0:
        F0_plot.append(0)
        F1_plot.append(0)
        F2_plot.append(0)
        F9_plot.append(0)
    else:
        F0_plot.append(F0(d[i]))
        F1_plot.append(F1(d[i]))
        F2_plot.append(F2(d[i]))
        F9_plot.append(F9(d[i]))

fig, ax1 = plt.subplots() 

ax1.plot(d, F0_plot, color='mediumseagreen', label="n=0") 
ax1.plot(d, F1_plot, color='turquoise', label="n=1") 
ax1.plot(d, F2_plot, color='dodgerblue', label="n=2") 
ax1.plot(d, F9_plot, color='mediumslateblue', label="n=9") 
ax1.set_title("Figure 6: Force Profile as a Function of Radius for \nDifferent n's") 
ax1.set_xlabel("Distance to the center of the Earth (m)") 
ax1.set_ylabel("Force profile (N)") 
ax1.legend();
```

Figure 6: This graph models the force that is applied to an object for different density distributions as they get further from the center of the Earth.

### 3. 

```python
alph = 0 
gam = 2

def Eq5_0(t, b): 
    y, v = b

    dydt = v 
    dvdt = -F0(y)+alph*v**gam
    return [dydt, dvdt]
def Eq5_1(t, b): 
    y, v = b

    dydt = v 
    dvdt = -F1(y)+alph*v**gam
    return [dydt, dvdt]
def Eq5_2(t, b): 
    y, v = b

    dydt = v 
    dvdt = -F2(y)+alph*v**gam
    return [dydt, dvdt]
def Eq5_9(t, b): 
    y, v = b

    dydt = v 
    dvdt = -F9(y)+alph*v**gam
    return [dydt, dvdt]

t0, tf = 0, 10000
t_eval = np.linspace(t0, tf, 50000) 

y0 = Re
v0 = 0
ic = [y0, v0] 

def stop(t, s): 
    y, v = s 
    return v 
stop.terminal = False

sol0 = solve_ivp(fun=Eq5_0, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop) 
sol1 = solve_ivp(fun=Eq5_1, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)
sol2 = solve_ivp(fun=Eq5_2, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)
sol9 = solve_ivp(fun=Eq5_9, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)

pos0 = sol0.y[0] 
vel0 = sol0.y[1] 
t_vals0 = sol0.t
pos1 = sol1.y[0] 
vel1 = sol1.y[1] 
t_vals1 = sol1.t
pos2 = sol2.y[0] 
vel2 = sol2.y[1] 
t_vals2 = sol2.t
pos9 = sol9.y[0] 
vel9 = sol9.y[1] 
t_vals9 = sol9.t

fig, ax2 = plt.subplots(4, 1, figsize=(10, 15)) 
 
ax2[0].plot(t_vals0, pos0, label='Position', color='blue') 
ax3 = ax2[0].twinx() 
ax3.plot(t_vals0, vel0, label='Velocity', color='green') 
ax2[0].set_title("Figure 7.1: (n=0)")
ax2[0].set_xlabel("Time (s)") 
ax2[0].set_ylabel("Position (m)", color='blue') 
ax3.set_ylabel("Velocity (m/s)", color='green') 

ax2[1].plot(t_vals1, pos1, label='Position', color='blue') 
ax3 = ax2[1].twinx() 
ax3.plot(t_vals1, vel1, label='Velocity', color='green')  
ax2[1].set_title("Figure 7.2: (n=1)")
ax2[1].set_xlabel("Time (s)") 
ax2[1].set_ylabel("Position (m)", color='blue') 
ax3.set_ylabel("Velocity (m/s)", color='green') 

ax2[2].plot(t_vals2, pos2, label='Position', color='blue') 
ax3 = ax2[2].twinx() 
ax3.plot(t_vals2, vel2, label='Velocity', color='green') 
ax2[2].set_title("Figure 7.3: (n=2)")
ax2[2].set_xlabel("Time (s)") 
ax2[2].set_ylabel("Position (m)", color='blue') 
ax3.set_ylabel("Velocity (m/s)", color='green') 

ax2[3].plot(t_vals9, pos9, label='Position', color='blue') 
ax3 = ax2[3].twinx() 
ax3.plot(t_vals9, vel9, label='Velocity', color='green') 
ax2[3].set_title("Figure 7.4: (n=9)")
ax2[3].set_xlabel("Time (s)") 
ax2[3].set_ylabel("Position (m)", color='blue') 
ax3.set_ylabel("Velocity (m/s)", color='green') 

fig.tight_layout() 
```

Figure 7: Each of these graphs contain the plots of the distance between to the center of the Earth and the velocity of an object falling through the Earth at different density distributions. 

### 4. 

```python
def stop(t, s): 
    y, v = s 
    return y 
stop.terminal=True

sol0 = solve_ivp(fun=Eq5_0, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop) 
sol1 = solve_ivp(fun=Eq5_1, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)
sol2 = solve_ivp(fun=Eq5_2, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)
sol9 = solve_ivp(fun=Eq5_9, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)

print(f" For density constant n=0, it takes {sol0.t_events[0][0]} seconds to reach the center of the Earth, with a speed of {np.abs(sol0.y_events[0][0][1])} m/s towards the center.\n")
print(f" For density constant n=1, it takes {sol1.t_events[0][0]} seconds to reach the center of the Earth, with a speed of {np.abs(sol1.y_events[0][0][1])} m/s towards the center.\n")
print(f" For density constant n=2, it takes {sol2.t_events[0][0]} seconds to reach the center of the Earth, with a speed of {np.abs(sol2.y_events[0][0][1])} m/s towards the center.\n")
print(f" For density constant n=9, it takes {sol9.t_events[0][0]} seconds to reach the center of the Earth, with a speed of {np.abs(sol9.y_events[0][0][1])} m/s towards the center.\n")
```
