## Part 3: The Coriolis Force

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp 
```

```python
G = 6.6743*10**-11 # gravitational constant [m^3/kg/s^2]
g0 = 9.81 # gravitational acceleration [m/s^2]
Me = 5.972*10**24 # mass of the earth [kg] 
Mm = 7.35*10**22 # mass of the moon [kg] 
Re = 6378.1e3 # radius of the earth [m] 
Rm = 1738.1e3 # radius of the moon [m] 
w = 7.272*10**-5 # earth's rotation rate at the equator [rad/s] 
```

### 1. 

```python
alp = 0
gam = 2 

def g(y): 
    gr = g0*(y/Re)
    return gr
    
def Eq3(t, vec): 
    x, y, vx, vy = vec 
    dxdt = vx 
    dydt = vy 
    dvxdt = 2*w*vy
    dvydt = -g(y)+alp*vy**gam-2*w*vx 
    return dxdt, dydt, dvxdt, dvydt  
```

### 2. 

```python
depth = 4*10**3 
a = g0
t = np.sqrt(depth/(0.5*a))

t0, tf = 0, t+5
t_eval = np.linspace(t0, tf, 1000) 

x0 = 0
y0 = Re 
vx0 = 0 
vy0 = 0 
ic=[x0, y0, vx0, vy0] 


sol = solve_ivp(fun=Eq3, t_span=(t0, tf), y0=ic, t_eval=t_eval)

xpos = sol.y[0] 
ypos = sol.y[1]
xvel = sol.y[2] 
yvel = sol.y[3] 
t_vals = sol.t

fig, ax = plt.subplots() 
ax.scatter(ypos[::30], xpos[::30], color='blue') 
ax.axvline(x=Re-4000, color='black', ls='--', label='Bottom of the shaft') 
ax.set_title("Figure 3.1: Transverse Position of the Test Mass with Coriolis Force, g(r), \nand no Drag. ") 
ax.set_xlabel("Distance to the center of the Earth (m)") 
ax.set_ylabel("X position (m)")
ax.legend(); 
```

Figure 3.1: This graph represents a sideways view of the path of the falling mass, taking into account the Coriolis force, changing g, and no drag. The x-axis represents the distance to the center of the Earth, which decreases towards the left, while the y-axis represents the sideways position of the test mass relative to its dropping point. It starts at zero, then starts moving further away from its original x position as it falls. The graph shows that the mass would reach the bottom shifted 5 meters to away from its original x position. 

### 3. 

```python
fig, ax1 = plt.subplots() 
ax1.scatter(ypos[::30], xpos[::30], color='blue') 
ax1.plot([Re, Re-4000], [2.5, 2.5], color='black', ls='--', label='Walls of the shaft') 
ax1.plot([Re, Re-4000], [-2.5, -2.5], color='black', ls='--') 
ax1.plot([Re-4000, Re-4000], [2.5, -2.5], color='black', ls='--') 
ax1.set_title("Figure 3.2: Transverse Position of the Test Mass inside the mineshaft \nwith Coriolis Force, g(r), and no Drag.") 
ax1.set_xlabel("Distance to the center of the Earth (m)") 
ax1.set_ylabel("X position (m)")
ax1.legend(); 
```

Figure 3.2: This figure is similar to Figure 3.1, but it includes the walls of the mineshaft, which is 5 meters wide. The graph shows that the test mass is dropped from the middle of the shaft, and would hit the wall before it reaches the bottom of the shaft due to the Coriolis Force. 

### 4. 

```python
vterm = vterm = 50 #m/s
alph = g0/vterm**gam
alp = alph

t0, tf = 0, t+70

sol = solve_ivp(fun=Eq3, t_span=(t0, tf), y0=ic, t_eval=t_eval)

xdrag = sol.y[0]
ydrag = sol.y[1] 
vxdrag = sol.y[2] 
vydrag = sol.y[3]
t_vals = sol.t

fig, ax2 = plt.subplots()
ax2.scatter(ypos[::30], xpos[::30], color='blue', label='Without drag') 
ax2.scatter(ydrag[::30], xdrag[::30], color='navy', label='With drag') 
ax2.plot([Re, Re-4000], [2.5, 2.5], color='black', ls='--', label='Walls of the shaft') 
ax2.plot([Re, Re-4000], [-2.5, -2.5], color='black', ls='--') 
ax2.plot([Re-4000, Re-4000], [2.5, -2.5], color='black', ls='--') 
ax2.legend() 
ax2.set_title("Figure 3.3: Transverse Position of the Test Mass inside the mineshaft \nwith Coriolis Force, g(r), and Drag.") 
ax2.set_xlabel("Distance to the center of the Earth (m)") 
ax2.set_ylabel("X position (m)"); 
```

Figure 3.3: This graph is similar to Figure 3.2, but it also shows the plot of the position of the test mass when taking drag into account. The graph shows that, when drag is added to the equation, the test mass hits the wall of the mineshaft much earlier than if drag were not present.

```python
alp = 0
def stop (t, s):
    x, y, vx, vy = s
    return x+2.5
    
stop.terminal = True
sol = solve_ivp(fun=Eq3, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)
tevent = sol.t_events[0][0]
yevent = sol.y_events[0][0][1]
print(f"When neglecting drag, the test mass will hit the wall of the shaft at {tevent} seconds and {Re-yevent} meters.") 

alp = alph
def stop (t, s):
    x, y, vx, vy = s
    return x+2.5
    
stop.terminal = True
sol = solve_ivp(fun=Eq3, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)
tevent = sol.t_events[0][0]
yevent = sol.y_events[0][0][1]
print(f"When accounting for drag, the test mass will hit the wall of the shaft at {tevent} seconds and {Re-yevent} meters.") 
```
