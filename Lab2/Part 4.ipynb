## Part 4: An infinitely deep mine

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp 
```

```python
G = 6.6743*10**-11 # gravitational constant [m^3/kg/s^2]
g0 = 9.81 # gravitational acceleration [m/s^2]
Me = 5.972*10**24 # mass of the earth [kg] 
Mm = 7.35*10**22 # mass of the moon [kg] 
Re = 6378.1e3 # radius of the earth [m] 
Rm = 1738.1e3 # radius of the moon [m] 
w = 7.272*10**-5 # earth's rotation rate at the equator [rad/s] 
```

### 1. 

```python
x = 2*Re
a = g0
t = np.sqrt(x/(0.5*a)) 

def g(y): 
    gr = g0*(y/Re)
    return gr
    
alph = 0 
gam = 2

def Eq4(t, b): 
    y, v = b

    dydt = v 
    dvdt = -g(y)+alph*v**gam
    return [dydt, dvdt]

t0, tf = 0, 8000
t_eval = np.linspace(t0, tf, 10000) 

y0 = Re
v0 = 0
ic = [y0, v0] 

sol = solve_ivp(fun=Eq4, t_span=(t0, tf), y0 = ic, t_eval = t_eval)

pos = sol.y[0] 
vel = sol.y[1] 
t_vals = sol.t 

fig, ax = plt.subplots() 
ax.plot(t_vals, pos, color='blue') 
ax.set_ylim(-1.1*Re, 1.1*Re) 
ax1 = ax.twinx() 
ax1.plot(t_vals, vel, label='Velocity', color='green') 
ax.axhline(y=0, color='gold', ls='--', label="Earth's center") 
ax.axhline(y=-Re, color='skyblue', ls='--', label="Earth's surface") 
ax.legend() 
ax.axhline(y=Re, color='skyblue', ls='--') 
ax.set_title("Figure 4: Position and Velocity of the Test Mass \nThrough an Infinitely Deep Mine")
ax.set_xlabel("Time (s)") 
ax.set_ylabel("Distance to the center of the Earth (m)", color='blue') 
ax1.set_ylabel("Velocity (m/s)", color='green');
```

Figure 4: This graph shows the plots of the distance of the test mass to the center of the Earth, as well as the velocity, if it were dropped down a theoretical mine that goes from one side of the Earth, through the center, and out the opposite end. The graph shows that the mass would fall all the way through to the other side, then fall back to the original side, then fall to the opposite side again. 

### 2.  

```python
def stop(t, s): 
    y, v = s 
    return v 

stop.terminal = False 
sol = solve_ivp(fun=Eq4, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)

event = sol.t_events[0][1] 

print(f"It will take the object {event} seconds to reach the other side.") 

def stop(t, s): 
    y, v = s 
    return y
    
stop.terminal = True
sol = solve_ivp(fun=Eq4, t_span=(t0, tf), y0=ic, t_eval=t_eval, events=stop)

t_event = sol.t_events[0][0] 
y_event = sol.y_events[0][0][1]

print(f"It will take the object {t_event} seconds to reach the center of the Earth, and it will be falling at {y_event} m/s as it passes through.")
```

### 3. 

```python
orb_v = np.sqrt(G*Me/Re) 
orb_t = np.pi*2*Re/orb_v

print(f'Given the orbital speed, {orb_v} m/s, the orbital period would be {orb_t} seconds, which is {orb_t/t_event} times the "crossing time".') 
```
